```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import matplotlib
from astropy.table import Table

matplotlib.rcParams.update({'font.size':18})
matplotlib.rcParams.update({'font.family':'serif'})
```

# Make demo figure

```python
# how to manually import code based on its path
import sys
sys.path.append('../ingot/')

from ingot import FeH
```

```python
df = pd.read_csv('../ingot/ingot_data.csv.gz')
```

```python
feh, feherr = FeH(df['w1']-df['w2'], df['phot_g_mean_mag']-df['Jmag'], df['M_G'], use_MG=True)
```

```python
plt.figure(figsize=(7,8))
plt.scatter(df['w1']-df['w2'],
            df['phot_g_mean_mag']-df['Jmag'],
            s=1, alpha=0.75, c=feh, cmap=plt.cm.coolwarm)

cb = plt.colorbar()
cb.set_label('[Fe/H]')
cb.set_clim(-1,0.5)
plt.xlim(-0.2,0.3)
plt.ylim(0.7, 3.1)
plt.xlabel('W1 - W2')
plt.ylabel('$G-J$ (mag)')
```

```python
plt.figure(figsize=(7,8))
plt.scatter(df['w1']-df['w2'],
            df['phot_g_mean_mag']-df['Jmag'],
            s=1, alpha=0.75, c=feherr, cmap=plt.cm.coolwarm)

cb = plt.colorbar()
cb.set_label('[Fe/H] Err')
cb.set_clim(0.10, 0.12)
plt.xlim(-0.2,0.3)
plt.ylim(0.7, 3.1)
plt.xlabel('W1 - W2')
plt.ylabel('$G-J$ (mag)')
```
